```python
#Fix the following Python function that is intended to calculate the factorial of a number: `def factorial(n): result = 1; for i in range(1, n): result = result * i; return result`. Test it using `factorial(5)` and explain why the output is incorrect. Identify the logical bug (for example, whether it is an off-by-one error) and clearly explain why it occurs. Then provide a corrected version of the function that returns 120 for `factorial(5)`, and improve it so that it properly handles edge cases such as zero, negative numbers, and non-integer inputs, including appropriate error handling and comments.
#The original function has an off-by-one error in the loop. The loop runs from 1 to n-1, which means it does not include n itself in the multiplication. For example, when `n` is 5, the loop multiplies 1 * 2 * 3 * 4, which results in 24 instead of the correct factorial value of 120 (which is 1 * 2 * 3 * 4 * 5).
# Fix the following Python function that is intended to calculate the factorial of a number:
#   def factorial(n):
#       result = 1
#       for i in range(1, n):
#           result = result * i
#       return result
#
# Test it using factorial(5) and explain why the output is incorrect.
# Identify the logical bug (for example, whether it is an off‑by‑one error) and
# clearly explain why it occurs. Then provide a corrected version of the function
# that returns 120 for factorial(5), and improve it so that it properly handles
# edge cases such as zero, negative numbers, and non‑integer inputs, including
# appropriate error handling and comments.
#
# The original function has an off‑by‑one error in the loop.  `range(1, n)` goes
# from 1 up to n‑1, so the factor `n` is never multiplied in.  For n = 5 the
# product is 1*2*3*4 = 24 instead of 1*2*3*4*5 = 120.

def factorial(n):
    """
    Compute the factorial of a non‑negative integer n.

    Parameters
    ----------
    n : int
        Non‑negative integer whose factorial is required.

    Returns
    -------
    int
        n!

    Raises
    ------
    TypeError
        If n is not an integer.
    ValueError
        If n is negative.
    """
    if not isinstance(n, int):
        raise TypeError("Input must be an integer")
    if n < 0:
        raise ValueError("Input must be a non-negative integer")

    # by definition 0! = 1
    if n == 0:
        return 1

    result = 1
    # include n itself by iterating to n + 1
    for i in range(1, n + 1):
        result *= i
    return result

# tests / demonstration
print(f"factorial(5) = {factorial(5)}")   # should print 120
print(f"factorial(0) = {factorial(0)}")   # should print 1

# the following lines, if uncommented, show the error handling:
# factorial(-1)        # raises ValueError
# factorial(3.5)       # raises TypeError
```
