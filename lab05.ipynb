```python
import requests

def get_weather(latitude, longitude):
    """
    Fetch weather data for given coordinates using Open-Meteo API.

    Args:
        latitude (float): Location latitude
        longitude (float): Location longitude

    Returns:
        dict: Weather data including temperature and wind speed
    """
    url = "https://api.open-meteo.com/v1/forecast"

    params = {
        "latitude": latitude,
        "longitude": longitude,
        "current": "temperature_2m,weather_code,wind_speed_10m",
        "temperature_unit": "fahrenheit"
    }

    try:
        response = requests.get(url, params=params)
        response.raise_for_status()
        return response.json()

    except requests.exceptions.RequestException as e:
        print(f"Error fetching weather data: {e}")
        return None


if __name__ == "__main__":
    # Example: New York City coordinates
    weather = get_weather(latitude=40.7128, longitude=-74.0060)

    if weather:
        current = weather.get("current", {})
        print(f"Temperature: {current.get('temperature_2m')} °F")
        print(f"Wind Speed: {current.get('wind_speed_10m')} mph")
```

```python
import json
import os
import bcrypt


def save_user_data(name, email, password, filename="users.json"):
    """
    Save user data to a JSON file with hashed password.
    """

    users = []

    # Load existing users if file exists
    if os.path.exists(filename):
        with open(filename, "r") as f:
            users = json.load(f)

    # Hash the password
    hashed_password = bcrypt.hashpw(
        password.encode("utf-8"),
        bcrypt.gensalt()
    ).decode("utf-8")

    # Add new user
    users.append({
        "name": name,
        "email": email,
        "password": hashed_password
    })

    # Save to file
    with open(filename, "w") as f:
        json.dump(users, f, indent=2)

    print(f"User {name} saved successfully!")


def load_users(filename="users.json"):
    """Load all users from file."""

    if os.path.exists(filename):
        with open(filename, "r") as f:
            return json.load(f)
    return []


def verify_password(stored_hash, password):
    """
    Verify a password against its hashed value.
    """
    return bcrypt.checkpw(
        password.encode("utf-8"),
        stored_hash.encode("utf-8")
    )


# Example usage
if __name__ == "__main__":
    save_user_data("John Doe", "john@example.com", "securepass123")
    save_user_data("Jane Smith", "jane@example.com", "password456")

    users = load_users()

    print("\nStored users:")
    for user in users:
        print(f"- {user['name']} ({user['email']})")

    # Verify password
    print("\nPassword verification:")
    print(
        verify_password(users[0]["password"], "securepass123")
    )
```

```python
#4.
#quicksort

def quicksort(arr):
    # Base case: if array has 0 or 1 elements, it's already sorted
    if len(arr) <= 1:
        return arr
    
    # Choose a pivot (here, the last element)
    pivot = arr[-1]
    
    # Partition step: divide elements into two lists
    left = [x for x in arr[:-1] if x <= pivot]  # elements smaller or equal to pivot
    right = [x for x in arr[:-1] if x > pivot]  # elements greater than pivot
    
    # Recursively sort left and right, then combine with pivot
    return quicksort(left) + [pivot] + quicksort(right)


# Example usage
print("QuickSort:", quicksort([64, 34, 25, 12, 22, 11, 90]))

#bubble sort
def bubblesort(arr):
    n = len(arr)
    # Outer loop: repeat passes through the list
    for i in range(n):
        # Inner loop: compare adjacent elements
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                # Swap if elements are in wrong order
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr


# Example usage
print("BubbleSort:", bubblesort([64, 34, 25, 12, 22, 11, 90]))
```

```python
#5.# Sample dataset: movies with genres and ratings
movies = [
    {"title": "Inception", "genre": "Sci-Fi", "rating": 8.8},
    {"title": "Interstellar", "genre": "Sci-Fi", "rating": 8.6},
    {"title": "The Dark Knight", "genre": "Action", "rating": 9.0},
    {"title": "Parasite", "genre": "Thriller", "rating": 8.6},
    {"title": "The Shawshank Redemption", "genre": "Drama", "rating": 9.3},
    {"title": "Avengers: Endgame", "genre": "Action", "rating": 8.4},
]

def recommend(user_pref_genre, min_rating=8.5):
    recommendations = []
    
    for movie in movies:
        # Check if movie matches user preference
        if movie["genre"] == user_pref_genre and movie["rating"] >= min_rating:
            reason = f"Recommended because it's a {movie['genre']} movie with a high rating of {movie['rating']}."
            recommendations.append((movie["title"], reason))
    
    # If no matches found, suggest top-rated movies in other genres
    if not recommendations:
        top_movies = sorted(movies, key=lambda x: x["rating"], reverse=True)[:2]
        for movie in top_movies:
            reason = f"Suggested because it's one of the highest-rated movies overall ({movie['rating']})."
            recommendations.append((movie["title"], reason))
    
    return recommendations


# Example usage
user_genre = "Sci-Fi"
results = recommend(user_genre)

for title, reason in results:
    print(f"{title} → {reason}")
#Explanation: The recommendation system filters movies based on the user's preferred genre and a minimum rating threshold. If no movies match the criteria, it suggests top-rated movies from other genres. Each recommendation includes a reason for why it was suggested, providing transparency to the user.
```
