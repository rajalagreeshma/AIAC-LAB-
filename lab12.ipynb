```python
# Task-1
# Write a program to manage student records with Name, Roll Number, and CGPA.
# mplement Quick Sort and Merge Sort to sort students in descending order of CGPA.
# Include functionality to measure and compare execution time for both algorithms on large datasets.
# Also, display the top 10 students with highest CGPA.
from heapq import merge
import time
class Student:
    def __init__(self, name, roll_number, cgpa):
        self.name = name
        self.roll_number = roll_number
        self.cgpa = cgpa

    def __str__(self):
        return f"{self.name} (Roll No: {self.roll_number}, CGPA: {self.cgpa})"
def quick_sort(students):
    if len(students) <= 1:
        return students
    pivot = students[len(students) // 2].cgpa
    left = [x for x in students if x.cgpa > pivot]
    middle = [x for x in students if x.cgpa == pivot]
    right = [x for x in students if x.cgpa < pivot]
    return quick_sort(left) + middle + quick_sort(right)
def merge_sort(students):
    if len(students) <= 1:
        return students
    mid = len(students) // 2
    left_half = merge_sort(students[:mid])
    right_half = merge_sort(students[mid:])
    return merge(left_half, right_half)
def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i].cgpa > right[j].cgpa:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result
def main():
    students = [
        Student("Alice", 1, 3.8),
        Student("Bob", 2, 3.5),
        Student("Charlie", 3, 3.9),
        Student("David", 4, 3.6),
        Student("Eve", 5, 3.7),
        Student("Frank", 6, 3.4),
        Student("Grace", 7, 3.9),
        Student("Heidi", 8, 3.2),
        Student("Ivan", 9, 3.8),
        Student("Judy", 10, 3.5)
    ]
    # Measure execution time for Quick Sort
    start_time = time.time()
    sorted_students_quick = quick_sort(students)
    end_time = time.time()
    print(f"Quick Sort Execution Time: {end_time - start_time:.6f} seconds")
    # Measure execution time for Merge Sort
    start_time = time.time()
    sorted_students_merge = merge_sort(students)
    end_time = time.time()
    print(f"Merge Sort Execution Time: {end_time - start_time:.6f} seconds")
    # Display top 10 students with highest CGPA
    print("\nTop 10 Students with Highest CGPA:")
    for student in sorted_students_quick[:10]:
        print(student)
if __name__ == "__main__":
    main()
```


Explanation: The program defines a Student class to represent student records, including their name, roll number, and CGPA. It implements both Quick Sort and Merge Sort algorithms to sort the students in descending order of CGPA. The execution time for both sorting algorithms is measured and printed. Finally, the program displays the top 10 students with the highest CGPA.

```python
# Task-2
# Write a Python program to implement Bubble Sort.
# Add detailed inline comments explaining each step, including comparisons, swapping of elements, passes through the list, and termination condition.
# Also include a clear time complexity analysis for best, average, and worst cases.
def bubble_sort(arr):
    n = len(arr)
    # Traverse through all elements in the list
    for i in range(n):
        # Initialize a flag to check if any swapping occurs
        swapped = False
        # Last i elements are already in place, no need to check them
        for j in range(0, n - i - 1):
            # Compare the adjacent elements
            if arr[j] > arr[j + 1]:
                # If the element found is greater than the next element, swap them
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        # If no swapping occurred, the list is already sorted, so we can break out of the loop
        if not swapped:
            break
    return arr
# Time Complexity Analysis:
# Best Case: O(n) - This occurs when the input list is already sorted. The algorithm makes one pass through the
# list to check if it is sorted, and since no swaps are needed, it terminates early.
# Average Case: O(n^2) - This occurs when the input list is in random order.
# The algorithm will need to make multiple passes through the list, and in each pass, it may need to compare and swap elements.
# Worst Case: O(n^2) - This occurs when the input list is sorted in reverse order.
# The algorithm will need to make n-1 passes through the list, and in each pass, it will need to compare and swap elements, resulting in n*(n-1)/2 comparisons and swaps.
# Example usage
if __name__ == "__main__":
    arr = [64, 34, 25, 12, 22, 11, 90]
    print("Original array:", arr)
    sorted_arr = bubble_sort(arr)
    print("Sorted array:", sorted_arr)
def sum_of_squares(numbers):
    total = 0
    for num in numbers:
        total += num ** 2
    return total
```

```python
#Task-3
# Write a Python program implementing Quick Sort and Merge Sort using recursion.
# Add docstrings explaining recursion and time complexities.
# Generate random, sorted, and reverse-sorted lists.
# Measure execution time for both algorithms.
# Print execution times and first 10 sorted elements for comparison.
import random
import time
def quick_sort(arr):
    """
    Quick Sort using recursion.
    Steps:
    1. Choose a pivot.
    2. Divide array into elements < pivot, == pivot, > pivot.
    3. Recursively sort left and right parts.
    """

    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]

    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)

def merge_sort(arr):
    """
    Merge Sort using recursion.
    Steps:
    1. Divide array into halves.
    2. Recursively sort both halves.
    3. Merge sorted halves.
    """

    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2

    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)

def merge(left, right):
    merged = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1

    merged.extend(left[i:])
    merged.extend(right[j:])
    return merged

# MAIN TESTING
if __name__ == "__main__":

    # Generate data
    random_list = [random.randint(1, 1000) for _ in range(1000)]
    sorted_list = sorted(random_list)
    reverse_sorted_list = sorted(random_list, reverse=True)

    # ---------------- QUICK SORT ----------------
    print("===== QUICK SORT =====")

    start = time.time()
    qs_random = quick_sort(random_list.copy())
    print("Random List Time:", time.time() - start)

    start = time.time()
    qs_sorted = quick_sort(sorted_list.copy())
    print("Sorted List Time:", time.time() - start)

    start = time.time()
    qs_reverse = quick_sort(reverse_sorted_list.copy())
    print("Reverse Sorted List Time:", time.time() - start)

    # ---------------- MERGE SORT ----------------
    print("\n===== MERGE SORT =====")

    start = time.time()
    ms_random = merge_sort(random_list.copy())
    print("Random List Time:", time.time() - start)

    start = time.time()
    ms_sorted = merge_sort(sorted_list.copy())
    print("Sorted List Time:", time.time() - start)

    start = time.time()
    ms_reverse = merge_sort(reverse_sorted_list.copy())
    print("Reverse Sorted List Time:", time.time() - start)

    # Show sample output instead of full list
    print("\nSample Sorted Output (First 10 elements):")
    print("Quick Sort:", qs_random[:10])
    print("Merge Sort:", ms_random[:10])
```

Explanation: The program defines two sorting functions, `quick_sort` and `merge_sort`, which implement the Quick Sort and Merge Sort algorithms using recursion. Each function includes a docstring explaining the steps of the algorithm and its time complexity. The program generates three types of lists: random, sorted, and reverse-sorted. It then measures and prints the execution time for both sorting algorithms on each type of list. Finally, it displays the first 10 sorted elements from the random list for both sorting methods to compare their outputs.

```python
#4 Build an Inventory Management System in Python.
# Each product should have: product_id, name, price, quantity.
# Implement efficient search:
# - Search by product ID using Binary Search
# - Search by name using Linear Search
# Implement sorting:
# - Sort by price using Quick Sort
# - Sort by quantity using Merge Sort
# Add clear comments explaining logic.
# Use simple code suitable for students.
# -------- Product Data --------
products = [
    [101, "Laptop", 50000, 10],
    [102, "Phone", 20000, 25],
    [103, "Tablet", 15000, 15],
    [104, "Monitor", 12000, 8],
    [105, "Keyboard", 2000, 50]
]

# -------- Binary Search (by ID) --------
def binary_search(products, target_id):
    low = 0
    high = len(products) - 1

    while low <= high:
        mid = (low + high) // 2

        if products[mid][0] == target_id:
            return products[mid]
        elif products[mid][0] < target_id:
            low = mid + 1
        else:
            high = mid - 1

    return "Product not found"

# -------- Linear Search (by Name) --------
def linear_search(products, name):
    for product in products:
        if product[1].lower() == name.lower():
            return product
    return "Product not found"

# -------- Quick Sort (by Price) --------
def quick_sort(products):
    if len(products) <= 1:
        return products

    pivot = products[len(products)//2][2]

    left = [x for x in products if x[2] < pivot]
    middle = [x for x in products if x[2] == pivot]
    right = [x for x in products if x[2] > pivot]

    return quick_sort(left) + middle + quick_sort(right)

# -------- Merge Sort (by Quantity) --------
def merge_sort(products):
    if len(products) <= 1:
        return products

    mid = len(products)//2
    left = merge_sort(products[:mid])
    right = merge_sort(products[mid:])

    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i][3] < right[j][3]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

# -------- Testing --------

# Sort by ID for Binary Search
products.sort(key=lambda x: x[0])

print("Search by ID:", binary_search(products, 103))
print("Search by Name:", linear_search(products, "Phone"))
print("\nSorted by Price:")
print(quick_sort(products))
print("\nSorted by Quantity:")
print(merge_sort(products))
```

Explanation: The program defines an inventory management system with a list of products, where each product has an ID, name, price, and quantity. It implements binary search to find products by their ID, linear search to find products by their name, quick sort to sort products by price, and merge sort to sort products by quantity. Each function is designed to be simple and easy to understand for students, with clear comments explaining the logic.

```python
# 5 Build a Python program to simulate stock market data.
# Each stock should have: symbol, opening price, closing price.
# Calculate percentage change for each stock.
# Sort stocks based on percentage change using Heap Sort.
# Implement fast search using Hash Map (dictionary) for stock symbols.
# Compare performance with Python's sorted() and dictionary lookup.
# Print results clearly and use simple code with comments.

import random
import time
import heapq

# -------- Generate Stock Data --------
def generate_stocks(n):
    stocks = []
    for i in range(n):
        symbol = "STK" + str(i)
        open_price = random.randint(100, 500)
        close_price = open_price + random.randint(-50, 50)

        change = ((close_price - open_price) / open_price) * 100

        stocks.append([symbol, open_price, close_price, round(change, 2)])

    return stocks


# -------- Heap Sort (by % change) --------
def heap_sort(stocks):
    heap = []

    # Push all elements into heap
    for stock in stocks:
        heapq.heappush(heap, stock[3])  # use % change

    sorted_list = []
    while heap:
        val = heapq.heappop(heap)
        
        # find stock with this change
        for stock in stocks:
            if stock[3] == val:
                sorted_list.append(stock)
                break

    return sorted_list


# -------- Hash Map Search --------
def create_hash_map(stocks):
    stock_dict = {}
    for stock in stocks:
        stock_dict[stock[0]] = stock
    return stock_dict


def search_stock(stock_dict, symbol):
    return stock_dict.get(symbol, "Stock not found")


# -------- Main --------
n = 1000
stocks = generate_stocks(n)

# Heap Sort Time
start = time.time()
heap_sorted = heap_sort(stocks)
print("Heap Sort Time:", time.time() - start)

# Built-in Sort Time
start = time.time()
builtin_sorted = sorted(stocks, key=lambda x: x[3])
print("Built-in Sort Time:", time.time() - start)

# Hash Map Creation
stock_map = create_hash_map(stocks)

# Search Time
start = time.time()
print("Search Result:", search_stock(stock_map, "STK10"))
print("Search Time:", time.time() - start)
```
